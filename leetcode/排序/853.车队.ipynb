**方法一：排序**

```python

class Solution(object):
    def carFleet(self, target, position, speed):
        cars = sorted(zip(position, speed))
        times = [float(target - p) / s for p, s in cars] # 计算车到终点的时间
        ans = 0
        while len(times) > 1:
            lead = times.pop()# 判断后面的车是否能追上前面的车
            if lead < times[-1]: ans += 1  # 可以追上，与前车速度保持一致
            else: times[-1] = lead # 追不上邻近的，保留该车

        return ans + bool(times) # remaining car is fleet (if it exists)

```
