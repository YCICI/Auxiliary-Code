**2.1 思路**

* 先按原点距离排序，再返回前k个值

**2.2 代码尝试**

```python
class Solution:
    def kClosest(self, points: List[List[int]], K: int) -> List[List[int]]:
        
        result=sorted(points,key = lambda x:x[0]**2+x[1]**2)
        return(result[:K])
    
# 解答正确
```

**2.3 正确代码解析**

**方法1** 先排序再输出距离最近的K个点
